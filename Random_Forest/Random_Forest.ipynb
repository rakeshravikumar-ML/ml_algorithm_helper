# Step 24: Feature Removal Analysis - Impact on Model Performance



## Feature Removal Analysis - Impact on Model Performance

### Breaking Down the Code:

```python
#HYPER PARAMETER TUNING USING GRIDSERACH
rf = RandomForestClassifier(random_state = 42)

np.random.seed(42)
start = time.time()

param_grid = {'bootstrap': [True, False],
              'max_features': ['auto', 'sqrt', 'log2', None],
              'criterion': ['gini', 'entropy']}

hyper_fit = GridSearchCV(rf,
                         cv = 5,
                         param_grid = param_grid,
                         n_jobs = -1)

hyper_fit.fit(x_train, y_train)

print('Bst Parameter using grid serach:', hyper_fit.best_params_)

end = time.time()

print('Time taken for grid serach:', end-start)
```

- **Line-by-line breakdown:**
  - Explanation for each function and parameter.


````
#HYPER PARAMETER TUNING USING GRIDSERACH
rf = RandomForestClassifier(random_state = 42)

np.random.seed(42)
start = time.time()

param_grid = {'bootstrap': [True, False],
              'max_features': ['auto', 'sqrt', 'log2', None],
              'criterion': ['gini', 'entropy']}

hyper_fit = GridSearchCV(rf,
                         cv = 5,
                         param_grid = param_grid,
                         n_jobs = -1)

hyper_fit.fit(x_train, y_train)

print('Bst Parameter using grid serach:', hyper_fit.best_params_)

end = time.time()

print('Time taken for grid serach:', end-start)
````

# Step 26: Optimizing Random Forest - Reducing Memory & Computation



## Optimizing Random Forest - Reducing Memory & Computation

### Breaking Down the Code:

```python
final_model.set_params(warm_start = True, oob_score = True)

min_estimator = 15
max_estimator = 1000

error_rate = {}

for trees in range(min_estimator, max_estimator+1):
  final_model.set_params(n_estimators = trees)
  final_model.fit(x_train, y_train)

  oob_error = 1-final_model.oob_score_

```

- **Line-by-line breakdown:**
  - Explanation for each function and parameter.


````
final_model.set_params(warm_start = True, oob_score = True)

min_estimator = 15
max_estimator = 1000

error_rate = {}

for trees in range(min_estimator, max_estimator+1):
  final_model.set_params(n_estimators = trees)
  final_model.fit(x_train, y_train)

  oob_error = 1-final_model.oob_score_
  error_rate[trees] = oob_error
````
